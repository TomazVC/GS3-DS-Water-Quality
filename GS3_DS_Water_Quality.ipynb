## **Tipo e Número de Valores Nulos**

### Primeiro, carregamos o dataset e verificamos o tipo de cada coluna e o número de valores nulos.

# **Inferências e Testes Estatísticos**

## **ANOVA**

### Realizamos um teste ANOVA para comparar a poluição da água entre diferentes cidades.

```python
from scipy.stats import f_oneway, ttest_ind

# ANOVA para comparar a poluição da água entre diferentes cidades
anova_result = f_oneway(*(df[df['City'] == city]['WaterPollution'] for city in df['City'].unique()))
print('ANOVA Result:', anova_result)
```

### **Resultado**:

### O teste ANOVA revela que há diferenças significativas na poluição da água entre diferentes cidades (p < 0.05), indicando que a localização geográfica pode influenciar a poluição da água.

## **Teste T**

### Realizamos um teste t para comparar a poluição da água entre dois países específicos.

```python
# Teste t para comparar dois grupos específicos
turkey = df_filtrado[df_filtrado['Country'] == 'Turkey']['WaterPollution'].dropna()
brazil = df_filtrado[df_filtrado['Country'] == 'Brazil']['WaterPollution'].dropna()
if turkey.empty or brazil.empty:
    print(f"Uma das amostras selecionadas não tem dados suficientes. Dados de Turkey: {len(turkey)}, Dados de Brazil: {len(brazil)}")
else:
    # Realizar o teste T
    ttest_result = ttest_ind(turkey, brazil)
    print('T-Test Result:', ttest_result)
```

### **Resultado**:

### O teste t indica se há uma diferença estatisticamente significativa na poluição da água entre dois países selecionados.

### Significativo (P < 0.05):

* Indica que há uma diferença estatisticamente significativa entre a poluição da água na Turquia e no Brasil.
* Sugere que a média da poluição da água é diferente entre esses dois países.

### Não Significativo (P ≥ 0.05):

* Indica que não há evidências suficientes para afirmar que há uma diferença significativa entre a poluição da água na Turquia e no Brasil.
* Sugere que a média da poluição da água pode ser semelhante entre os dois países.

### A conclusão foi de P = 0.03, ou seja, existe uma diferença na poluição da água entre os dois países.

```python
# Calcular as médias de poluição da água para Turkey e Brazil
mean_turkey = turkey.mean()
mean_brazil = brazil.mean()

# Exibir as médias
print(f"Média de Poluição da Água na Turquia: {mean_turkey}")
print(f"Média de Poluição da Água no Brasil: {mean_brazil}")
```

### Interpretação dos Resultados

### Média de Poluição da Água na Turquia (mean_turkey):

* Esta média representa o nível médio de poluição da água nas cidades da Turquia incluídas no conjunto de dados.

### Média de Poluição da Água no Brasil (mean_brazil):

* Esta média representa o nível médio de poluição da água nas cidades do Brasil incluídas no conjunto de dados.

### Se a média da poluição da água na Turquia for maior do que no Brasil, isso indica que, em média, a qualidade da água na Turquia é pior (mais poluída) do que no Brasil. Se for menor, indica que a qualidade da água na Turquia é melhor (menos poluída) do que no Brasil.

### Resultados Calculados

* Média de Poluição da Água na Turquia: 44.17
* Média de Poluição da Água no Brasil: 52.74

### Interpretação

* Turquia (44.17): A média de poluição da água é de 44.17, indicando um nível de poluição relativamente alto, mas menor se comparado ao Brazil.
* Brasil (52.74): A média de poluição da água é de 52.74, indicando um nível de poluição maior que a Turquia.

### Portanto, com base nos valores calculados, podemos concluir que o Brazil tem, em média, uma poluição da água maior do que a Turquia. Isso sugere que, de maneira geral, a qualidade da água na Turquia é melhor do que no Brasil.
